```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#load the file
df = pd.read_csv(r"Model_Performance_Comparison.csv", encoding='ISO-8859-1')

df.head()
```

```python
# Extract mean and SD for R²
df['R²_mean'] = df['R²'].str.split('±').str[0].astype(float)
df['R²_sd'] = df['R²'].str.split('±').str[1].astype(float)

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# Plot using ax
bars = ax.bar(df['Model'], df['R²_mean'], color='grey')

# Use ax methods
ax.set_xlabel('Model', fontsize=10)
ax.set_ylabel('R²', fontsize=10)
ax.set_title('Model Comparison - R²', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
ax.set_ylim(0, 1)

# Add text inside bars
for bar, value in zip(bars, df['R²_mean']):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        value / 2,
        f"{value:.2f}",
        ha='center',
        va='center',
        fontsize=10,
        fontweight='bold',
        color='white'
    )

plt.tight_layout()
plt.savefig("05.Model_comparison_r2.png", dpi=600)
plt.show()
```
